```python
import os
import csv
import numpy as np
```

```python
def read_file(run,dataset_name,influence_type,at_type):
        
    if at_type == '1':
        parent_dir = '/home/amanmoha/influence_attack/results/actual/DT/modules/'
    elif at_type == '2':
        parent_dir = '/home/amanmoha/influence_attack/results/actual/DT/modules_ref_tuning/'
    
    directory = 'run'+run+'/'+dataset_name+'/'+influence_type
    path = os.path.join(parent_dir,directory)
    
    with open(path+'/metrics.txt','r') as f:
        lines = f.readlines()
    f.close()
    
    data = []
    for line in lines:
        d_ = line.split('\t')
        data.append(np.round(float(d_[-1].strip(' \n')),3))
    return data
```

```python
run = '1'
dataset_names = ['bcw','pima_diabetes','hepatitis','german_credit']
influence_types = ['SHAP','DP_impurity','DP_importance']
at_types = ['2']
filename = "new_results_DT.csv"

'''
at_type = 1 -- modules
at_type = 2 -- modules ref tuning f1
'''

header1 = ['','Train Accuracy','','Test Accuracy','','Generalization Error','','Score F1-Attack Accuracy','','','',
           'Score F1-Attack Precision','','','']
header2 = ['Dataset','Our','SM','Our','SM','Our','SM','SHAP','Impurity','Importance','Base Acc','SHAP','Impurity','Importance','Base Pre']

data_csv = []
for dataset_name in dataset_names:
    
    if dataset_name == 'bcw':
        SM_data = [0.987,0.944,0.514,0.523,0.545,0.528] 
    elif dataset_name == 'pima_diabetes':
        SM_data = [0.789,0.756,0.517,0.519,0.537,0.561]
    elif dataset_name == 'hepatitis':
        SM_data = [0.997,0.81,0.605,0.618,0.562,0.609]
    elif dataset_name == 'german_credit':
        SM_data = [0.937,0.701,0.623,0.622,0.603,0.637]

    data_metrics = []
    data_acc = []
    data_pre = []
    data_rec = []
    data_base = []

    for at_type in at_types:
        for influence_type in influence_types:
            data = read_file(run,dataset_name,influence_type,at_type)
            data_metrics.append(data[:2])
            data_acc.append(data[2])
            data_pre.append(data[3])
            data_rec.append(data[4])
            data_base.append(data[6:-1])
            
#     data_ = [dataset_name]+[data_metrics[0][0]]+[SM_data[0]]+[data_metrics[0][1]]+[SM_data[1]]+[data_metrics[0][0]-data_metrics[0][1]]+[SM_data[0]-SM_data[1]]+data_acc+SM_data[2:4]+[data_base[0][0]]+data_pre+[data_base[0][1]]+SM_data[4:6]+data_rec+[data_base[0][2]]
    data_ = [dataset_name]+[data_metrics[0][0]]+[SM_data[0]]+[data_metrics[0][1]]+[SM_data[1]]+[data_metrics[0][0]-data_metrics[0][1]]+[SM_data[0]-SM_data[1]]+data_acc+[data_base[0][0]]+data_pre+[data_base[0][1]]
    data_csv.append(data_) 
    
with open(filename, 'w') as csvfile:    
    csvwriter = csv.writer(csvfile)   
    csvwriter.writerow(header1) 
    csvwriter.writerow(header2) 
    csvwriter.writerows(data_csv) 
    
csvfile.close()
```
